```python
import pickle
import numpy as np
import os
```

```python
def load_data_pkl():
    data = []
    project = os.getcwd().split('/')[3] + "_pseudo"
    case = "557af118-bf38-460f-81fa-b95aa746295d"
    data_pkl = "/work/jack91630/DATA/PROCESSED_DATA/CASE_UUID/{}/{}.pkl".format(project, case)
    with open(data_pkl, 'rb') as f:
        while True:
            try:
                data.append(pickle.load(f))
            except EOFError:
                break

    # data 0: tumor, 1: normal, 2: background

    return data
```

```python
from itertools import combinations_with_replacement
import numpy as np
import random
c = list(combinations_with_replacement("ABCDEFGHIJKLMNO",3))
```

```python
for i in c[0]:
    print(ord(i) - 65)
```

```python
for x in range(len(c)):
    c[x] = list(c[x])
    for y in range(len(c[x])):
        c[x][y] = ord(c[x][y]) - 65
```

```python
weight = [i for i in range(len(c))]
```

```python
random.choices(c, weights = weight)
```
